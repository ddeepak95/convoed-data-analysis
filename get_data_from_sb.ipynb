```python
import os
from general_functions import get_supabase_client

client = get_supabase_client()
```

```python
import json
def save_to_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f, indent=4)
```

```python
def get_class_details(class_code: str) -> str:
    response = client.table("classes").select("*").eq("class_id", class_code).execute()
    return response
```

```python
folder_name = "data/021826"

# create folder if it doesn't exist
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
class_code = "epakC0LO"

class_details = get_class_details(class_code)
class_uid = class_details.data[0]["id"]

save_to_json(class_details.data, os.path.join(folder_name, "class_details.json"))
```

```python
class_contents = client.table("content_items").select("*").eq("class_id", class_uid).execute()
print(class_contents)
```
