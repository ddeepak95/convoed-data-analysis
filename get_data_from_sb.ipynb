```python
import os
from general_functions import get_supabase_client

client = get_supabase_client()
```

```python
import json
def save_to_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f, indent=4)

def read_json(filename):
    with open(filename, 'r') as f:
        return json.load(f)
```

```python
class_code = "epakC0LO"
folder_name = "data/021826"

# create folder if it doesn't exist
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
class_details = client.table("classes").select("*").eq("class_id", class_code).execute()
class_uid = class_details.data[0]["id"]

save_to_json(class_details.data, os.path.join(folder_name, "class_details.json"))
```

```python
class_contents = client.table("content_items").select("*").eq("class_id", class_uid).execute()
save_to_json(class_contents.data, os.path.join(folder_name, "class_contents.json"))
```

```python
class_groups = client.table("class_groups").select("*").eq("class_id", class_uid).execute()
save_to_json(class_groups.data, os.path.join(folder_name, "class_groups.json"))
```

```python
class_assignments = client.table("assignments").select("*").eq("class_id", class_uid).execute()
save_to_json(class_assignments.data, os.path.join(folder_name, "class_assignments.json"))
```

```python
class_learning_contents = client.table("learning_contents").select("*").eq("class_id", class_uid).execute()
save_to_json(class_learning_contents.data, os.path.join(folder_name, "class_learning_contents.json"))
```

```python
class_quizzes = client.table("quizzes").select("*").eq("class_id", class_uid).execute()
save_to_json(class_quizzes.data, os.path.join(folder_name, "class_quizzes.json"))
```

```python
class_surveys = client.table("surveys").select("*").eq("class_id", class_uid).execute()
save_to_json(class_surveys.data, os.path.join(folder_name, "class_surveys.json"))
```

```python
quiz_submissions = client.table("quiz_submissions").select("*").eq("class_id", class_uid).execute()
save_to_json(quiz_submissions.data, os.path.join(folder_name, "quiz_submissions.json"))
```

```python
activity_logs = client.table("activity_logs").select("*").eq("class_id", class_uid).execute()
save_to_json(activity_logs.data, os.path.join(folder_name, "activity_logs.json"))
```

```python
activity_events = client.table("activity_events").select("*").eq("class_id", class_uid).execute()
save_to_json(activity_events.data, os.path.join(folder_name, "activity_events.json"))
```

```python
assignments = read_json(os.path.join(folder_name, "class_assignments.json"))

list_of_assignment_codes = [a["assignment_id"] for a in assignments]

assignment_submissions = client.table("submissions").select("*").in_("assignment_id", list_of_assignment_codes).execute()
save_to_json(assignment_submissions.data, os.path.join(folder_name, "assignment_submissions.json"))
```

```python
class_contents = read_json(os.path.join(folder_name, "class_contents.json"))
class_content_ids = [c["id"] for c in class_contents]

content_completions = client.table("student_content_completions").select("*").in_("content_item_id", class_content_ids).execute()
save_to_json(content_completions.data, os.path.join(folder_name, "content_completions.json"))
```
